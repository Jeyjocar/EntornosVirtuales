```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,4,3])
plt.show() 
```

```python
"""Code permite agreagar nuevas lneas de cdigo"""

import numpy as np

x = np.linspace(0,2,100)

plt.plot(x,x,label = "lineal")

plt.plot(x,x**2,label = "cuadrtica")
plt.plot(x,x**3, label = "cubica")
plt.xlabel("Mi Label X")
plt.ylabel("Mi Label Y")
plt.title("MI Grfica")
plt.legend()
plt.show()




```

```python
"""Ejercicio POO"""

x = np.linspace(0,2,100)
fig,ax = plt.subplots(figsize=(5,2.7), layout = "constrained")
ax.plot(x,x, label = "lineal")
ax.plot(x,x**2,label = "cuadrtica")
ax.plot(x,x**3, label = "cubica")
ax.set_xlabel("Mi Label X")
ax.set_ylabel("Mi Label Y")
ax.set_title("Mi Grfica")
ax.legend()
ax.show()
```

```python
ejex = np.linspace(0.1,7)
ejey = np.sin(ejex)
plt.plot(ejex,ejey,linewidth=4,linestyle=(0,(4,3,15,1)))
plt.show()
```

```python
ejex = np.linspace(0.1,6)
ejey = np.sin(ejex)
fig,ax = plt.subplots()

ax.stem(ejex,ejey,linefmt=("y",":"),markerfmt="D",basefmt="C8--")
ax.spines["left"].set_color("blue")
ax.spines["top"].set_color("red")
ax.spines["bottom"].set_color("green")
ax.spines["right"].set_color("orange")
ax.spines["top"].set_linewidth(4)
ax.spines["top"].set_alpha(0.8) #set_alpha = sombreado
ax.spines["top"].set_linestyle(":")
ax.grid(axis = "x", visible=True,linewidth=2,linestyle=(0,(4,3)),color="black")
ax.show()
```
